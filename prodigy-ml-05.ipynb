```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import warnings
warnings.filterwarnings('ignore')

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision
from torchvision import transforms, models
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
import time
from tqdm import tqdm
import cv2

# Check for GPU availability
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"CUDA Version: {torch.version.cuda}")

class FoodDataset(Dataset):
    def __init__(self, data_dir, split='train', transform=None, limit_classes=None):
        """
        Custom Dataset for Food-101
        
        Args:
            data_dir: Path to the food-101 directory
            split: 'train' or 'test'
            transform: Image transformations
            limit_classes: Limit number of classes for faster training (None for all)
        """
        self.data_dir = data_dir
        self.split = split
        self.transform = transform
        
        # Load class names
        with open(os.path.join(data_dir, 'meta', 'classes.txt'), 'r') as f:
            self.classes = [line.strip() for line in f.readlines()]
        
        if limit_classes:
            self.classes = self.classes[:limit_classes]
        
        self.class_to_idx = {cls: idx for idx, cls in enumerate(self.classes)}
        
        # Load train/test split
        split_file = os.path.join(data_dir, 'meta', f'{split}.txt')
        self.image_paths = []
        self.labels = []
        
        with open(split_file, 'r') as f:
            for line in f.readlines():
                img_path = line.strip()
                class_name = img_path.split('/')[0]
                
                if class_name in self.classes:
                    full_path = os.path.join(data_dir, 'images', f'{img_path}.jpg')
                    if os.path.exists(full_path):
                        self.image_paths.append(full_path)
                        self.labels.append(self.class_to_idx[class_name])
        
        print(f"Loaded {len(self.image_paths)} images for {split} set")
    
    def __len__(self):
        return len(self.image_paths)
    
    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        label = self.labels[idx]
        
        try:
            image = Image.open(img_path).convert('RGB')
            if self.transform:
                image = self.transform(image)
            return image, label
        except Exception as e:
            print(f"Error loading image {img_path}: {e}")
            # Return a black image if loading fails
            if self.transform:
                return self.transform(Image.new('RGB', (224, 224), (0, 0, 0))), label
            return Image.new('RGB', (224, 224), (0, 0, 0)), label

class FoodRecognitionModel(nn.Module):
    def __init__(self, num_classes, pretrained=True):
        super(FoodRecognitionModel, self).__init__()
        
        # Use ResNet50 as backbone
        self.backbone = models.resnet50(pretrained=pretrained)
        
        # Replace the final layer
        num_features = self.backbone.fc.in_features
        self.backbone.fc = nn.Sequential(
            nn.Dropout(0.5),
            nn.Linear(num_features, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, num_classes)
        )
    
    def forward(self, x):
        return self.backbone(x)

def create_calorie_database():
    """
    Create a simple calorie database for Food-101 classes
    Note: These are approximate values per 100g
    """
    calorie_data = {
        'apple_pie': 237, 'baby_back_ribs': 292, 'baklava': 307, 'beef_carpaccio': 114,
        'beef_tartare': 196, 'beet_salad': 43, 'beignets': 347, 'bibimbap': 121,
        'bread_pudding': 219, 'breakfast_burrito': 206, 'bruschetta': 194, 'caesar_salad': 158,
        'cannoli': 248, 'caprese_salad': 159, 'carrot_cake': 349, 'ceviche': 89,
        'cheese_plate': 368, 'cheesecake': 321, 'chicken_curry': 119, 'chicken_quesadilla': 215,
        'chicken_wings': 203, 'chocolate_cake': 371, 'chocolate_mousse': 168, 'churros': 117,
        'clam_chowder': 103, 'club_sandwich': 282, 'crab_cakes': 169, 'creme_brulee': 338,
        'croque_madame': 258, 'cup_cakes': 305, 'deviled_eggs': 124, 'donuts': 452,
        'dumplings': 41, 'edamame': 122, 'eggs_benedict': 301, 'escargots': 90,
        'falafel': 333, 'filet_mignon': 277, 'fish_and_chips': 265, 'foie_gras': 462,
        'french_fries': 365, 'french_onion_soup': 57, 'french_toast': 289, 'fried_calamari': 175,
        'fried_rice': 163, 'frozen_yogurt': 159, 'garlic_bread': 350, 'gnocchi': 131,
        'greek_salad': 150, 'grilled_cheese_sandwich': 295, 'grilled_salmon': 231, 'guacamole': 160,
        'gyoza': 64, 'hamburger': 295, 'hot_and_sour_soup': 91, 'hot_dog': 290,
        'huevos_rancheros': 233, 'hummus': 166, 'ice_cream': 207, 'lasagna': 135,
        'lobster_bisque': 104, 'lobster_roll_sandwich': 227, 'macaroni_and_cheese': 164, 'macarons': 407,
        'miso_soup': 40, 'mussels': 86, 'nachos': 346, 'omelette': 154,
        'onion_rings': 411, 'oysters': 68, 'pad_thai': 153, 'paella': 172,
        'pancakes': 227, 'panna_cotta': 236, 'peking_duck': 337, 'pho': 46,
        'pizza': 266, 'pork_chop': 231, 'poutine': 365, 'prime_rib': 291,
        'pulled_pork_sandwich': 233, 'ramen': 436, 'ravioli': 175, 'red_velvet_cake': 478,
        'risotto': 166, 'samosa': 308, 'sashimi': 127, 'scallops': 137,
        'seaweed_salad': 45, 'shrimp_and_grits': 151, 'spaghetti_bolognese': 151, 'spaghetti_carbonara': 174,
        'spring_rolls': 140, 'steak': 271, 'strawberry_shortcake': 227, 'sushi': 143,
        'tacos': 226, 'takoyaki': 112, 'tiramisu': 240, 'tuna_tartare': 144,
        'waffles': 291
    }
    return calorie_data

def visualize_data_distribution(dataset, title):
    """Visualize class distribution"""
    class_counts = {}
    for _, label in dataset:
        class_name = dataset.classes[label]
        class_counts[class_name] = class_counts.get(class_name, 0) + 1
    
    plt.figure(figsize=(15, 8))
    classes = list(class_counts.keys())[:20]  # Show top 20 classes
    counts = [class_counts[cls] for cls in classes]
    
    plt.bar(range(len(classes)), counts)
    plt.title(f'{title} - Top 20 Classes Distribution')
    plt.xlabel('Food Classes')
    plt.ylabel('Number of Images')
    plt.xticks(range(len(classes)), classes, rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

def visualize_sample_images(dataset, num_samples=16):
    """Visualize sample images from dataset"""
    fig, axes = plt.subplots(4, 4, figsize=(12, 12))
    fig.suptitle('Sample Food Images', fontsize=16)
    
    for i in range(num_samples):
        img, label = dataset[i * len(dataset) // num_samples]
        
        # Convert tensor to numpy if needed
        if isinstance(img, torch.Tensor):
            # Denormalize if normalized
            img = img.permute(1, 2, 0)
            if img.max() <= 1:
                img = img * 255
            img = img.numpy().astype(np.uint8)
        
        row, col = i // 4, i % 4
        axes[row, col].imshow(img)
        axes[row, col].set_title(dataset.classes[label], fontsize=10)
        axes[row, col].axis('off')
    
    plt.tight_layout()
    plt.show()

def train_model(model, train_loader, val_loader, num_epochs=10, learning_rate=0.001):
    """Train the food recognition model"""
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.1)
    
    model = model.to(device)
    
    train_losses = []
    train_accuracies = []
    val_losses = []
    val_accuracies = []
    
    best_val_acc = 0.0
    
    print("Starting training...")
    
    for epoch in range(num_epochs):
        start_time = time.time()
        
        # Training phase
        model.train()
        running_loss = 0.0
        correct_predictions = 0
        total_samples = 0
        
        train_pbar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs} [Train]')
        for batch_idx, (data, targets) in enumerate(train_pbar):
            data, targets = data.to(device), targets.to(device)
            
            optimizer.zero_grad()
            outputs = model(data)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total_samples += targets.size(0)
            correct_predictions += (predicted == targets).sum().item()
            
            # Update progress bar
            train_pbar.set_postfix({
                'Loss': f'{loss.item():.4f}',
                'Acc': f'{100 * correct_predictions / total_samples:.2f}%'
            })
        
        train_loss = running_loss / len(train_loader)
        train_acc = 100 * correct_predictions / total_samples
        
        # Validation phase
        model.eval()
        val_running_loss = 0.0
        val_correct = 0
        val_total = 0
        
        with torch.no_grad():
            val_pbar = tqdm(val_loader, desc=f'Epoch {epoch+1}/{num_epochs} [Val]')
            for data, targets in val_pbar:
                data, targets = data.to(device), targets.to(device)
                outputs = model(data)
                loss = criterion(outputs, targets)
                
                val_running_loss += loss.item()
                _, predicted = torch.max(outputs.data, 1)
                val_total += targets.size(0)
                val_correct += (predicted == targets).sum().item()
                
                val_pbar.set_postfix({
                    'Loss': f'{loss.item():.4f}',
                    'Acc': f'{100 * val_correct / val_total:.2f}%'
                })
        
        val_loss = val_running_loss / len(val_loader)
        val_acc = 100 * val_correct / val_total
        
        # Store metrics
        train_losses.append(train_loss)
        train_accuracies.append(train_acc)
        val_losses.append(val_loss)
        val_accuracies.append(val_acc)
        
        # Update learning rate
        scheduler.step()
        
        epoch_time = time.time() - start_time
        
        print(f'Epoch {epoch+1}/{num_epochs}:')
        print(f'  Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%')
        print(f'  Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%')
        print(f'  Time: {epoch_time:.2f}s')
        print('-' * 50)
        
        # Save best model
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), 'best_food_model.pth')
    
    return {
        'train_losses': train_losses,
        'train_accuracies': train_accuracies,
        'val_losses': val_losses,
        'val_accuracies': val_accuracies,
        'best_val_acc': best_val_acc
    }

def plot_training_history(history):
    """Plot training history"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    # Plot losses
    ax1.plot(history['train_losses'], label='Train Loss', marker='o')
    ax1.plot(history['val_losses'], label='Validation Loss', marker='s')
    ax1.set_title('Model Loss')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend()
    ax1.grid(True)
    
    # Plot accuracies
    ax2.plot(history['train_accuracies'], label='Train Accuracy', marker='o')
    ax2.plot(history['val_accuracies'], label='Validation Accuracy', marker='s')
    ax2.set_title('Model Accuracy')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy (%)')
    ax2.legend()
    ax2.grid(True)
    
    plt.tight_layout()
    plt.show()

def evaluate_model(model, test_loader, class_names):
    """Evaluate model on test set"""
    model.eval()
    all_predictions = []
    all_targets = []
    
    with torch.no_grad():
        for data, targets in tqdm(test_loader, desc='Evaluating'):
            data, targets = data.to(device), targets.to(device)
            outputs = model(data)
            _, predicted = torch.max(outputs, 1)
            
            all_predictions.extend(predicted.cpu().numpy())
            all_targets.extend(targets.cpu().numpy())
    
    # Calculate metrics
    accuracy = accuracy_score(all_targets, all_predictions)
    report = classification_report(all_targets, all_predictions, 
                                 target_names=class_names, output_dict=True)
    
    return accuracy, report, all_predictions, all_targets

def plot_confusion_matrix(y_true, y_pred, class_names, top_n=20):
    """Plot confusion matrix for top N classes"""
    # Get top N classes by frequency
    unique, counts = np.unique(y_true, return_counts=True)
    top_classes_idx = unique[np.argsort(counts)[-top_n:]]
    
    # Filter data for top classes
    mask = np.isin(y_true, top_classes_idx)
    y_true_filtered = np.array(y_true)[mask]
    y_pred_filtered = np.array(y_pred)[mask]
    
    # Create mapping for filtered classes
    class_mapping = {old_idx: new_idx for new_idx, old_idx in enumerate(top_classes_idx)}
    y_true_mapped = [class_mapping[idx] for idx in y_true_filtered]
    y_pred_mapped = [class_mapping[idx] for idx in y_pred_filtered]
    
    top_class_names = [class_names[idx] for idx in top_classes_idx]
    
    cm = confusion_matrix(y_true_mapped, y_pred_mapped)
    
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=top_class_names,
                yticklabels=top_class_names)
    plt.title(f'Confusion Matrix - Top {top_n} Classes')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

def predict_with_calories(model, image_path, transform, class_names, calorie_db):
    """Predict food class and estimate calories"""
    model.eval()
    
    # Load and preprocess image
    image = Image.open(image_path).convert('RGB')
    original_image = np.array(image)
    
    input_tensor = transform(image).unsqueeze(0).to(device)
    
    with torch.no_grad():
        outputs = model(input_tensor)
        probabilities = torch.nn.functional.softmax(outputs[0], dim=0)
        top5_prob, top5_idx = torch.topk(probabilities, 5)
    
    # Get predictions
    predictions = []
    for i in range(5):
        class_name = class_names[top5_idx[i].item()]
        prob = top5_prob[i].item()
        calories = calorie_db.get(class_name, 'Unknown')
        predictions.append((class_name, prob, calories))
    
    return original_image, predictions

def create_prediction_visualization(image, predictions):
    """Create visualization for predictions"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Show original image
    ax1.imshow(image)
    ax1.set_title('Input Image')
    ax1.axis('off')
    
    # Show predictions
    classes = [pred[0].replace('_', ' ').title() for pred in predictions]
    probs = [pred[1] for pred in predictions]
    colors = plt.cm.viridis(np.linspace(0, 1, len(classes)))
    
    bars = ax2.barh(range(len(classes)), probs, color=colors)
    ax2.set_yticks(range(len(classes)))
    ax2.set_yticklabels(classes)
    ax2.set_xlabel('Confidence')
    ax2.set_title('Top 5 Predictions')
    ax2.set_xlim(0, 1)
    
    # Add probability text on bars
    for i, (bar, prob, pred) in enumerate(zip(bars, probs, predictions)):
        ax2.text(bar.get_width() + 0.01, bar.get_y() + bar.get_height()/2,
                f'{prob:.2%}\n({pred[2]} cal/100g)', 
                va='center', fontsize=9)
    
    plt.tight_layout()
    plt.show()

def generate_model_report(history, test_accuracy, classification_report, class_names):
    """Generate comprehensive model report"""
    print("="*80)
    print("FOOD RECOGNITION MODEL - COMPREHENSIVE REPORT")
    print("="*80)
    
    print(f"\n📊 DATASET INFORMATION:")
    print(f"   • Total Classes: {len(class_names)}")
    print(f"   • Model Architecture: ResNet50 with custom classifier")
    print(f"   • Device Used: {device}")
    
    print(f"\n🎯 TRAINING RESULTS:")
    print(f"   • Best Validation Accuracy: {history['best_val_acc']:.2f}%")
    print(f"   • Final Training Accuracy: {history['train_accuracies'][-1]:.2f}%")
    print(f"   • Final Validation Accuracy: {history['val_accuracies'][-1]:.2f}%")
    
    print(f"\n🔍 TEST SET PERFORMANCE:")
    print(f"   • Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
    
    # Top performing classes
    class_scores = {}
    for class_name in class_names:
        if class_name in classification_report:
            class_scores[class_name] = classification_report[class_name]['f1-score']
    
    top_classes = sorted(class_scores.items(), key=lambda x: x[1], reverse=True)[:10]
    
    print(f"\n🏆 TOP 10 PERFORMING CLASSES (F1-Score):")
    for i, (class_name, score) in enumerate(top_classes, 1):
        print(f"   {i:2d}. {class_name.replace('_', ' ').title():<25} {score:.3f}")
    
    print(f"\n📈 MODEL METRICS:")
    print(f"   • Macro Average Precision: {classification_report['macro avg']['precision']:.3f}")
    print(f"   • Macro Average Recall: {classification_report['macro avg']['recall']:.3f}")
    print(f"   • Macro Average F1-Score: {classification_report['macro avg']['f1-score']:.3f}")
    print(f"   • Weighted Average F1-Score: {classification_report['weighted avg']['f1-score']:.3f}")
    
    print(f"\n💡 CALORIE ESTIMATION:")
    print(f"   • Integrated calorie database with {len(create_calorie_database())} food items")
    print(f"   • Provides approximate calories per 100g for recognized foods")
    
    print(f"\n🚀 USAGE RECOMMENDATIONS:")
    print(f"   • Model works best with well-lit, clear food images")
    print(f"   • Consider image quality and angle for better predictions")
    print(f"   • Calorie estimates are approximate - consult nutrition labels for accuracy")
    
    print("="*80)

# Main execution function
def main():
    # Configuration
    DATA_DIR = '/kaggle/input/food101/food-101'  
    BATCH_SIZE = 32
    NUM_EPOCHS = 15
    LEARNING_RATE = 0.001
    LIMIT_CLASSES = 20
    
    # Data transformations
    train_transform = transforms.Compose([
        transforms.Resize((256, 256)),
        transforms.RandomCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(15),
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    test_transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    print("Loading datasets...")
    
    # Create datasets
    train_dataset = FoodDataset(DATA_DIR, split='train', transform=train_transform, limit_classes=LIMIT_CLASSES)
    test_dataset = FoodDataset(DATA_DIR, split='test', transform=test_transform, limit_classes=LIMIT_CLASSES)
    
    # Create validation split from training data
    train_size = int(0.8 * len(train_dataset))
    val_size = len(train_dataset) - train_size
    train_subset, val_subset = torch.utils.data.random_split(train_dataset, [train_size, val_size])
    
    # Create data loaders
    train_loader = DataLoader(train_subset, batch_size=BATCH_SIZE, shuffle=True, num_workers=4, pin_memory=True)
    val_loader = DataLoader(val_subset, batch_size=BATCH_SIZE, shuffle=False, num_workers=4, pin_memory=True)
    test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=4, pin_memory=True)
    
    print(f"Training samples: {len(train_subset)}")
    print(f"Validation samples: {len(val_subset)}")
    print(f"Test samples: {len(test_dataset)}")
    
    # Visualize data
    visualize_data_distribution(train_dataset, "Training Data")
    visualize_sample_images(train_dataset)
    
    # Create model
    model = FoodRecognitionModel(num_classes=len(train_dataset.classes))
    print(f"Created model with {sum(p.numel() for p in model.parameters())} parameters")
    
    # Train model
    history = train_model(model, train_loader, val_loader, num_epochs=NUM_EPOCHS, learning_rate=LEARNING_RATE)
    
    # Plot training history
    plot_training_history(history)
    
    # Load best model for evaluation
    model.load_state_dict(torch.load('best_food_model.pth'))
    
    # Evaluate on test set
    test_accuracy, classification_report_dict, predictions, targets = evaluate_model(model, test_loader, train_dataset.classes)
    
    # Plot confusion matrix
    plot_confusion_matrix(targets, predictions, train_dataset.classes)
    
    # Create calorie database
    calorie_db = create_calorie_database()
    
    # Generate comprehensive report
    generate_model_report(history, test_accuracy, classification_report_dict, train_dataset.classes)
    
    # Example prediction (you can modify the path)
    if os.path.exists('sample_food_image.jpg'):
        image, preds = predict_with_calories(model, 'sample_food_image.jpg', test_transform, 
                                           train_dataset.classes, calorie_db)
        create_prediction_visualization(image, preds)
    
    print("\n✅ Training and evaluation completed!")
    print("📁 Best model saved as 'best_food_model.pth'")
    
    return model, history, train_dataset.classes, calorie_db

# Example usage for prediction
def predict_food_image(model_path, image_path, classes, calorie_db):
    """
    Function to predict a single food image
    
    Args:
        model_path: Path to saved model
        image_path: Path to image to predict
        classes: List of class names
        calorie_db: Calorie database dictionary
    """
    # Load model
    model = FoodRecognitionModel(num_classes=len(classes))
    model.load_state_dict(torch.load(model_path))
    model = model.to(device)
    
    # Transform
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    # Predict
    image, predictions = predict_with_calories(model, image_path, transform, classes, calorie_db)
    create_prediction_visualization(image, predictions)
    
    return predictions

if __name__ == "__main__":
    # Run the main training pipeline
    model, history, classes, calorie_db = main()
    
    # Save class names and calorie database for future use
    import pickle
    with open('food_classes.pkl', 'wb') as f:
        pickle.dump(classes, f)
    
    with open('calorie_database.pkl', 'wb') as f:
        pickle.dump(calorie_db, f)
```
